# Data Visualization

### Histographs

```python
fig, axs = plt.subplots(2,3,figsize = (10,10))

axs[0,0].hist(df['Volume'])
axs[0,0].set_title('Volume')

axs[0,1].hist(df['Avg Price'])
axs[0,1].set_title('Avg Price')

axs[0,2].hist(df['Total Sales Value'])
axs[0,2].set_title('Total Sales Value')

axs[1,0].hist(df['Discount Rate (%)'])
axs[1,0].set_title('Discount Rate (%)')

axs[1,1].hist(df['Discount Amount'])
axs[1,1].set_title('Discount Amount')

axs[1,2].hist(df['Net Sales Value'])
axs[1,2].set_title('Net Sales Value')

plt.tight_layout()

plt.show()
```

### To get a rough idea on skewness, we can use scipy.stats module to measure skewness of each numerical column. If skewness > 0 it is a positive skewness and if skewness < 0 it is a negative skewness


```python
from scipy.stats import skew

columns = df[['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']]
for col in columns:
    print(f"{col}: {skew(df[col])}")
```

### Box Plot

#### With the help of box plots we can visualize outliers in the dataset. Below are the box plots for numerical columns. The small circles shown in the diagram are outliers

```python
fig,axs = plt.subplots(2,3, figsize=(15,10))
axs = axs.flatten()

for i,col in enumerate(columns):
    axs[i].boxplot(df[col])
    axs[i].set_title(col)
```

# Standardization of numerical variables

### We can convert the numerical columns of the dataset through standardization technique which helps the ML model to predict more accurate values. We can do this using mathematical formulae of mean and standard deviation or using scikit learn module in python

```python
from sklearn.preprocessing import StandardScaler

Scaler = StandardScaler()
columns = Scaler.fit_transform(columns)
columns = pd.DataFrame(columns, columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value'])
columns
```

### Box plot after standardizing the numerical columns

```python
fig,axs = plt.subplots(2,3, figsize=(15,10))
axs = axs.flatten()

for i,col in enumerate(columns):
    axs[i].boxplot(df[col])
    axs[i].set_title(col)
```

### Skewness after standardizing numerical columns
